```python
import pandas as pd

twitter_data = pd.read_csv('Data/Twitter_Clean.csv')
energy_data = pd.read_csv('Data/WorldEnergyConsumptionClean.csv')

twitter_data['country'] = twitter_data['country'].str.lower()
energy_data['country'] = energy_data['country'].str.lower()

# Merge the two datasets on 'year' and 'country'
merged_data = pd.merge(twitter_data, energy_data, how='inner', on=['year', 'country'])

merged_data.head(50)
```
