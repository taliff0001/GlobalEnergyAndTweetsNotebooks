```python
import pandas as pd


file_path = "Data/World Energy Consumption.csv"
# Load the dataset
energy_f = pd.read_csv(file_path)
```

```python
print("Current rows", len(energy_f))
energy_f.describe()
```

```python
# Filter the data for the years 2008 to 2019
filtered_year = energy_f[(energy_f['year'] >= 2008) & (energy_f['year'] <= 2019)]
#Remove ASEAN (ember)
filtered_ember = filtered_year[filtered_year['country'] != 'ASEAN (Ember)']
# Select the specific columns as requested
filtered_columns = filtered_ember[['country', 'year', 'renewables_share_energy', 'fossil_share_energy', 'primary_energy_consumption']]
# Remove rows where both 'renewables_share_energy' and 'fossil_share_energy' are null
filtered_data = filtered_columns.dropna(subset=['renewables_share_energy', 'fossil_share_energy'], how='all')
```

```python
print("Filtered data:", filtered_data)
```

```python
# View the filtered data for the USA without creating a new subset
print(filtered_data[filtered_data['country'] == 'United States'])
```

```python
filtered_data.to_csv('Data/OWID_clean.csv', index=False)
```
